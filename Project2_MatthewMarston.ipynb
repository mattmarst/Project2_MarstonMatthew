```python
import random

words = ["WHDSPQZ", "XHO", "TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]

# chooses random word from list and stores length of word
chosen_word = random.choice(words)
word_length = len(chosen_word)

# creates list with each letter of chosen word as element and list of letters used
word_letters = list(chosen_word)

# creates lives, used letters list, and initial current word list
lives = 10
correct_guesses = 0
letters_used = []
current_word = []
for i in range (word_length):
    current_word.append("-")

# for testing purposes
# print(chosen_word)
# print(word_length)

# While lives are greater than 0 or the word is not guessed, code executes.
while (lives > 0) and (correct_guesses < word_length):
    
    updated_word = " ".join(current_word)
    updated_letters = (" ").join(letters_used)
    
    # Tells users their lives, letters they have guessed, what the word looks like with guessed letters, and asks for 
    print(f"You have {lives} lives left and you have used these letters: {updated_letters} \nCurrent word: {updated_word}")
    guess = input("Guess a letter: ")
    print("\n")
    
    # tells user if they already guessed word
    if (guess.upper() in letters_used):
        print(f"Sorry, looks like you already guessed the letter {guess.upper()}.")
    
    # checks if guessed letter is in word and will display whether it is or not in next loop
    elif guess.upper() in word_letters:
        for i in range(word_length):
            if guess.upper() == word_letters[i]:
                current_word[i] = guess.upper()
                correct_guesses += 1
        letters_used.append(guess.upper())
    
    # if guess not in word, tells user and takes life away    
    else:
        print(f"Your letter, {guess.upper()} is not in the word.")
        lives -= 1
        letters_used.append(guess.upper())
    
# if user guesses all the letters in the word, it outputs statement below
if correct_guesses == word_length:
    print(f"Yay! You guessed the word {chosen_word}!")
    print(f"Decrypted word: {decrypt(chosen_word)}")

else:
    print("\nYou ran out of lives!")
```

```python
def decrypt(encrypted_word):
    
    cypher_list = list(encrypted_word)
    plain_list = []
    
    i = 0
    
    for i in range(len(cypher_list)):
        # takes cypher ascii value from letter
        cypher_ascii = ord(cypher_list[i])
        
        # if i is even, subtract 1 from cypher_ascii to get plain_ascii
        if i%2 == 0:
            plain_ascii = cypher_ascii - 1
        
        # if i is odd, add 1 to cypher_ascii to get plain_ascii
        else:
            plain_ascii = cypher_ascii + 1
        
        # converts ascii to char and adds to plaintext word
        plain_text = chr(plain_ascii)
        plain_list.append(plain_text)
    
    decrypted_word = "".join(plain_list)
    
    return decrypted_word
        
    
```
